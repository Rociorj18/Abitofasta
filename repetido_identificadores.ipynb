```python
from google.colab import files

# Función para verificar identificadores repetidos en el archivo FASTA
def check_duplicates(fasta_file):
    sequence_ids = set()
    duplicates = []

    for line in fasta_file:
        if line.startswith('>'):
            sequence_id = line.strip().split('|')[0][1:]
            if sequence_id in sequence_ids:
                duplicates.append(sequence_id)
            else:
                sequence_ids.add(sequence_id)

    return duplicates

# Cargar archivo desde el PC
uploaded = files.upload()

# Iterar sobre los archivos cargados
for filename in uploaded.keys():
    print(f'Archivo cargado: {filename}')

    # Leer el contenido del archivo
    with open(filename, 'r') as file:
        fasta_content = file.readlines()

    # Verificar identificadores duplicados
    duplicates = check_duplicates(fasta_content)

    # Mostrar resultados
    if duplicates:
        print('Se encontraron identificadores duplicados:')
        for duplicate in duplicates:
            print(duplicate)
    else:
        print('No se encontraron identificadores duplicados en el archivo.')
```

```python
# Función para verificar identificadores duplicados en varios archivos FASTA y ver cuales no
def check_duplicates_multiple(files_content):
    all_sequence_ids = set()
    duplicates = []

    for fasta_content in files_content:
        sequence_ids = set()

        for line in fasta_content:
            if line.startswith('>'):
                sequence_id = line.strip().split('|')[0][1:]
                if sequence_id in sequence_ids or sequence_id in all_sequence_ids:
                    duplicates.append(sequence_id)
                else:
                    sequence_ids.add(sequence_id)
                    all_sequence_ids.add(sequence_id)

    return duplicates, all_sequence_ids

# Cargar archivos desde el PC
uploaded = files.upload()

# Leer contenido de los archivos
files_content = []
for filename in uploaded.keys():
    print(f'Archivo cargado: {filename}')

    with open(filename, 'r') as file:
        fasta_content = file.readlines()
        files_content.append(fasta_content)

# Verificar identificadores duplicados entre archivos
duplicates, all_sequence_ids = check_duplicates_multiple(files_content)

# Identificar los identificadores únicos
unique_ids = all_sequence_ids - set(duplicates)

# Mostrar resultados
if unique_ids:
    print('Identificadores que no están duplicados:')
    for unique_id in unique_ids:
        print(unique_id)
else:
    print('Todos los identificadores están duplicados.')
```
